```python
cnt = 1
lines = []
with open('un.txt' , 'r+') as fp:
    line = fp.readline()
    while line:
        print("Line {}: {}".format(cnt, line.strip()))
        if cnt % 2 == 1:
            print ("this is it",line.split()[0][:-1])
            lines.append(line.split()[0].strip('\n')[:-1] + ' '+line.split()[1].lstrip('\n\r') +\
                        ': ' +'return ' +'"'+line.split()[1].lstrip('\n\r') +'"'+ ';')
        else:
            lines.append(line)
        line = fp.readline()
        cnt += 1
```

```python
with open('final.txt' , 'w') as fp:
    for line in lines:
        print("new line",line)
        fp.write(line)
        fp.write('\n')
```
